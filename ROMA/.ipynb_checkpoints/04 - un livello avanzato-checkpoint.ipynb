## Filtro della qualità delle sequenze

Discriminiamo dunque in base a una soglia minima di qualità sul carattere.

Es. 35

````python
for carattere in qualita:
    if codifica(carattere) > 35:
        print('Passato il filtro:', carattere)
````

`IF` è il costrutto Python per gestire decisioni basate su condizioni

````python
for carattere in qualita:
    if codifica(carattere) > 35:
        print('Passato il filtro:', carattere)
    else:
        print('Fallito il filtro:', carattere)
````

````python
somma = 0
for carattere in qualita:
    somma += codifica(carattere)  # equivalente a somma = somma + codifica
    
print("Somma della qualità:", somma)
print("Media:", somma/len(qualita))
````

Proviamo qualcosa di più complesso

````python
def filtra(qualita, soglia=35):  # ho messo il valore di 'default'
    
    somma = 0

    for carattere in qualita:
        somma += codifica(carattere)  # equivalente a somma = somma + codifica
    media = somma/len(qualita)
    return media > soglia
````

````python
filtra('ABCD')
````

````python
filtra('ABCDEFGHI')
````

````python
filtra('ABCD', 30)
````

````python
def leggi_quattro_righe(handler):
    next(handler)
    sequenza = next(handler)
    next(handler)
    qualita = next(handler)
    
    return sequenza, qualita
````

````python
file = 'data/raw.fastq'
with open(file) as handler:
    sequenza, qualita = leggi_quattro_righe(handler)
    print(sequenza)
    print(qualita)
````

## Ottimizzazione

Riccardo scopre di dover rendere l'esecuzione dei propri codici nei cluster del CINECA più veloce

````python
def filtra_lista(dati):
    lista = []
    for elemento in dati:
        if elemento > 5:
            lista.append(elemento)
    return lista
````

````python
filtra_lista([1,3,5,7,9])
````

````python
for elemento in filtra_lista([1,3,5,7,9]):
    print(elemento)
````

Ecco cosa accade con le liste:

````python
def generatore(dati):
    # lista = []
    for elemento in dati:
        if elemento > 5:
            # lista.append(elemento)
            yield elemento
    # return lista
````

````python
generatore([1,3,5,7,9])
````

````python
g = generatore([1,3,5,7,9])
````

````python
next(g)
````

## Riscontro rispetto al genoma di riferimento

Riccardo ha dei campioni di DNA di cui conosce le sequenze.

Quanto sono diverse da quelle del genoma umano "generico"?

Queste 'differenze' si chiamano **mutazioni**.

La domanda innanzitutto è:

dove si trovano posizionate le sequenze ottenute rispetto al genoma umano di riferimento?

Esistono dei programmi (pubblicati su articoli scientifici) che effettuano per noi questo compito difficile.

(es. `bwa` oppure `bowtie`)

CINECA ci supporta in questa fase computazionalmente dispendiosa.

Idealmente passi successivi 

- scrivere output filtrato
- copiare file su cluster cineca (oppure lavorare in python direttamente sul cluster)
- eseguire job di allineamento con qsub
- recuperare output o verificare eventuali errori
- abbiamo già esempio di output (SAM)
